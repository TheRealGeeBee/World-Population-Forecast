```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\GABBY\Downloads\archive.zip")
df.info()
```

```python
for item in df["Yearly Change"]:
    new_item = item.translate( { ord("%"): None } )
    df['Yearly Change'].replace(item, new_item, inplace=True)
print("Done!")
```

```python
df.isnull().sum()
```

```python
    columns_dict = {"Country (or dependency)":object,
    "Population (2020)":int,    
    "Yearly Change":float,  
    "Net Change":int,  
    "Density (P/Km²)":int,  
    "Land Area (Km²)":int,  
    "Migrants (net)":float,
    "Fert. Rate":object, 
     "Med. Age":object, 
    "Urban Pop %":object, 
    "World Share":object, }
    
    df = df.astype(columns_dict)
    df.info()
```

```python
df[:21]
```

```python
yearly_change_list = []
for values in df['Yearly Change']:
    values = values/100
    yearly_change_list.append(values)
```

```python
country_list = []
for country in df['Country (or dependency)']:
    country_list.append(country)
```

```python
last_year = []
current_year = []
for items in df["Population (2020)"]:
    last_year.append(items)
```

```python
year = 2021

while year < 2030:
    counter = 0
    for value in last_year:
        new_population = (value*yearly_change_list[counter]) + value
        current_year.append(new_population)
        counter += 1
    year += 1
    last_year = current_year
    current_year = []
print("Done!")
```

```python
last_year
```

```python
land_area = []
for item in df['Land Area (Km²)']:
    land_area.append(item)
```

```python
new_counter = 0
population_2020 = []
for item in df["Population (2020)"]:
    population_2020.append(item)
percent_change = []
for value in last_year:
    change = ((value-population_2020[last_year.index(value)])/population_2020[last_year.index(value)])*100
    percent_change.append(round(change, 2))
```

```python
data_dict = {'Country (or dependency)': country_list,
            "Population (2029)": last_year,
            'Land Area (Km²)': land_area,
            'Percentage Change in Population (from 2020)': percent_change
            }
```

```python
df = pd.DataFrame(data_dict)
df = df.sort_values("Population (2029)", ascending=False)
df['Percentage Change in Population (from 2020)'].max()
df[df['Percentage Change in Population (from 2020)'] == 40.37] #Niger is the country that would see the highest growth in population
```

```python
df['Percentage Change in Population (from 2020)'].min()
df[df['Percentage Change in Population (from 2020)'] == -20.16] #Puerto Rico would record the lowest population growth.
```

## Look at the table above, which represents the projected population for countries in 2029. Compare it with the one higher above and you would see that, in 2029, India could be the most populous country in the world. There are also other new comers on the top-twenty list, who were not there in 2020. Puerto Rico, a country in North America, would record the lowest growth. It would see a 20.16% decrease in its population.

```python
my_list = []
for items in df['Country (or dependency)']:
    my_list.append(items)
true_colour = []
for country in my_list:
    if len(country) == 3:
        true_colour.append("black")
    elif len(country) == 4:
        true_colour.append("gray")
    elif len(country) == 5:
        true_colour.append("green")
    elif len(country) == 6:
        true_colour.append("orange")
    elif len(country) == 7:
        true_colour.append("yellow")
    elif len(country) == 8:
        true_colour.append("blue")
    elif len(country) == 9:
        true_colour.append("turquoise")
    elif len(country) == 10:
        true_colour.append("maroon")
    else:
        true_colour.append("red")
```

```python
x_column = df['Land Area (Km²)']
y_column = df['Population (2029)']
clr = []
for items in x_column:
    if items < 100:
        clr.append('blue')
    elif items < 1000:
        clr.append('green')
    elif items < 10000:
        clr.append('gray')
    elif items < 100000:
        clr.append('yellow')
    elif items < 1000000:
        clr.append('black')
    elif items < 10000000:
        clr.append('maroon')
    elif items < 100000000:
        clr.append('magenta')
    elif items < 1000000000:
        clr.append('orange')
    else:
        clr.append('red')
print(clr)
```

```python
plt.scatter(x_column, y_column, c=true_colour)
plt.title("Population Of Countries In The World In 2029")
plt.xlabel("Country Land Area (in Km²)")
plt.ylabel("Population")
plt.xscale('log')
plt.yscale('log')
countries = df['Country (or dependency)']
counter = 0
while counter <= 233:
    plt.text(x_column[counter], y_column[counter], my_list[counter], fontsize=6)
    counter+= 1
plt.savefig("World_Population_2029.png")
plt.show()
```
